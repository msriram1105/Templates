```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense, Dropout
from keras.layers.normalization import BatchNormalization


#Initialising the CNN
classifier = Sequential()

# AlexNet Architecture
# 1st Convolutional Layer
classifier.add(Convolution2D(filters = 96, input_shape = (224, 224, 3), kernel_size = (11, 11), strides = (4, 4), padding = 'valid', activation = 'relu'))
# Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2), strides = (2, 2), padding = 'valid'))
#Batch Normalization
classifier.add(BatchNormalization())

#2nd Convolutional Layer
classifier.add(Convolution2D(filters = 256, kernel_size = (11, 11), strides = (1, 1), padding = 'valid', activation = 'relu'))
#Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2), strides = (2, 2), padding = 'valid'))
#Batch Normalization
classifier.add(BatchNormalization())

#3rd Convolutional Layer
classifier.add(Convolution2D(filters = 384, kernel_size = (3, 3), strides = (1, 1), padding = 'valid', activation = 'relu'))
#Batch Normalization
classifier.add(BatchNormalization())

#4th Convolutional layer
classifier.add(Convolution2D(filters = 384, kernel_size = (3, 3), strides = (1, 1), padding = 'valid', activation = 'relu'))
#Batch Normalization
classifier.add(BatchNormalization())

#5th Convolutional Layer
classifier.add(Convolution2D(filters = 256, kernel_size = (3, 3), strides = (1, 1), padding = 'valid', activation = 'relu'))
#Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2), strides = (2, 2), padding = 'valid'))
#Batch Normalization
classifier.add(BatchNormalization())

# Passing it to a Dense Layer
classifier.add(Flatten())
#1st Dense Layer
classifier.add(Dense(4096, input_shape = (224*224*3,), activation = 'relu'))
#Adding Dropout
classifier.add(Dropout(0.4))
#Batch Normalization
classifier.add(BatchNormalization())

#2nd Dense Layer
classifier.add(Dense(4096, activation = 'relu'))
#Adding Dropout
classifier.add(Dropout(0.4))
#Batch Normalization
classifier.add(BatchNormalization())

#Output Layer
classifier.add(Dense(output_dim = 17, activation = 'softmax'))
classifier.add(Dense(output_dim = 2, activation = 'sigmoid'))

classifier.summary()

#Compiling 
classifier.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Part 2 - Fitting the CNN to the images
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

training_set = train_datagen.flow_from_directory(
        'E:\\Deep_Learning_A_Z\\Volume 1 - Supervised Deep Learning\\Part 2 - Convolutional Neural Networks (CNN)\\Section 8 - Building a CNN\\dataset\\training_set',
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

test_set = test_datagen.flow_from_directory(
        'E:\\Deep_Learning_A_Z\\Volume 1 - Supervised Deep Learning\\Part 2 - Convolutional Neural Networks (CNN)\\Section 8 - Building a CNN\\dataset\\test_set',
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

classifier.fit(
        training_set,
        steps_per_epoch=8000,
        epochs=20,
        validation_data=test_set,
        validation_steps =2000)




```
